```python
import numpy as np
import pandas as pd
```

```python
data_dir = 'gc-mc/data/ml_100k/'
train = pd.read_csv(data_dir + 'u1.base', sep='\t', header=None)
test = pd.read_csv(data_dir + 'u1.test', sep='\t', header=None)
train.rename(columns={0: 'UserID', 1: 'MovieID', 2: 'Rating', 3: 'Timestamp'}, inplace=True)
test.rename(columns={0: 'UserID', 1: 'MovieID', 2: 'Rating', 3: 'Timestamp'}, inplace=True)

train.index += 1
test.index += 1

train.head()
```

```python
train.to_csv('train.csv', columns=['UserID', 'MovieID', 'Rating'], index_label='TrainDataID')
test.to_csv('test.csv', columns=['UserID', 'MovieID', 'Rating'], index_label='TestDataID')
```

```python
occupations = [line.rstrip('\n') for line in open(data_dir + 'u.occupation')]
genres = [line.split('|')[0] for line in open(data_dir + 'u.genre')]
```

```python
users = pd.read_csv(data_dir + 'u.user', sep='|', header=None)
users.rename(columns={0: 'UserID', 1: 'Age', 2: 'Gender', 3: 'Occupation', 4: 'Zip-code'}, inplace=True)
users.Occupation = users.Occupation.apply(lambda x: occupations.index(x))

users.head()
```

```python
users.to_csv('users.csv', columns=['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code'], index=None)
```

```python
movies = pd.read_csv(data_dir + 'u.item', sep='|', header=None, engine='python')
```

```python
def parse_movie(row):
    row['movieID'] = row[0]
    row['Title'] = row[1]
    row['Genres'] = []
    for i in range(5, 24):
        if row[i] == 1:
            row['Genres'].append(genres[i - 5])
    row['Genres'] = '|'.join(row['Genres'])
    return row

movies = movies.apply(parse_movie, axis=1)

movies.head()
```

```python
movies.to_csv('movies.csv', columns=['movieID', 'Title', 'Genres'], index=None)
```
